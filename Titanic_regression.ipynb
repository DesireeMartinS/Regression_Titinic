```python
dfTitanic1= dfTitanic.drop(['name','ticket','cabin','boat','home'],axis=1)
dfTitanic1.columns

```

# Converting obj -> int data type/ Clean nan
1.Identify list of unique values
2.Create a dictionary
3. Replace obj data type values with int data type values using dic
4. A function could also be used to replace values

```python
#sex Creade dic array(['female', 'male'])

sexNum={'female':1, 'male':0}
```

```python
#sex Replace values using dic

dfTitanic1= dfTitanic1.replace({'sex':sexNum})
```

```python
#embarked array(['S', 'C', nan, 'Q'])

embarkedNum={'S':1, 'C':2, 'Q':3}
```

```python
#embarked replace values using dic

dfTitanic1 = dfTitanic1.replace({'embarked':embarkedNum})
```

```python
#Status of nan values

dfTitanic1.isnull().sum()
```

```python
#Clean data
dfTitanic2= dfTitanic1.drop(['body'],axis=1)
#dfTitanic2.columns
```

```python
dfTitanic3= dfTitanic2.dropna()
```

```python
dfTitanic3.columns
```

```python
dfTitanic3.isnull().sum()
```

```python
#Drop dependable variable; 
from sklearn.linear_model import LinearRegression
X = dfTitanic3.drop(['survived'], axis = 1)

#assign linear regression function to a variable
lm = LinearRegression()
lm


```

```python
lm.fit(X, dfTitanic3.survived)
```

```python
print('Estimated intercept coefficient:', lm.intercept_)
```

```python
print('Number of coefficients:', len(lm.coef_))
```

```python
pd.DataFrame(list(zip(X.columns, lm.coef_)), columns = ['features', 'estimatedCoefficients'])
```
